#### For single CSV File 

```python
import csv

def read_csv_column(filename, column_index):
    data = []
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        for i, row in enumerate(csv_reader):
            if i > 0:  # Skip the first row
                if row[column_index] != 'Inf':
                    data.append(float(row[column_index]))  # Convert the entry to numeric
    return data

filename = 'aco_40.csv'
column_index = 2  # Index of the column you want to read (0-based indexing)

aco_40pc = read_csv_column(filename, column_index)
print(aco_40pc)
```

#### For Multiple CSV Files 

```python
import csv

def read_csv_column(filename, column_index):
    data = []
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            value = row[column_index]
            if value != 'inf':
                data.append(float(value))  # Convert the entry to numeric
    return data

def read_multiple_csv_files(filenames, column_index):
    all_data = []
    for filename in filenames:
        column_data = read_csv_column(filename, column_index)
        all_data.append(column_data)
    return all_data

filenames = ['ACO_results_nocongestion.csv', 'ACO_results_20congestion.csv', 'ACO_results_100congestion.csv']  # List of filenames
column_index = 2 # Index of the column you want to read (0-based indexing)

column_data_1, column_data_2, column_data_3 = read_multiple_csv_files(filenames, column_index)
all_column_data = read_multiple_csv_files(filenames, column_index)

print("Column 1 data:", column_data_1)
print("Column 2 data:", column_data_2)
print("Column 3 data:", column_data_3)

```

```python
import matplotlib.pyplot as plt

#data = [column_data_1, column_data_2, column_data_3, column_data_4]

# Creating box plots
plt.boxplot(all_column_data)

# Adding labels to the x-axis
plt.xticks([1, 2, 3], ['No Congestion', '20 % Congestion', '100 % Congestion'], rotation =45)

# Adding title and labels
plt.title('Box plots for column data')
plt.xlabel('Level of Congestion')
plt.ylabel('Time Cost')
```
